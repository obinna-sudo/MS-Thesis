```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import time 

current_time = time.time()
print("current time in seconds since the epoch: " ,current_time )
#loadDataset
fuzzy_data = pd.read_csv('../Multi_class_Fuzzy_Dos/fuzzing__data_original .csv')
dos_data = pd.read_csv('../Multi_class_Fuzzy_Dos/DoS_data_original .csv')
```

```python
fuzzy_data['Class_Instance'].value_counts()
```

```python
fuzzy_data
```

```python
fuzzy_data['Class_Instance'] = fuzzy_data['Class_Instance'].replace({'T': 'F', 'R': 'N'})
```

```python
fuzzy_data['Class_Instance'].value_counts()
```

```python
dos_data['Class_Instance'] = dos_data['Class_Instance'].replace({'T': 'D', 'R': 'N'})
```

```python
dos_data['Class_Instance'].value_counts()
```

```python
#concatenate Fuzzy and DoS attack Dataset
combined_data = pd.concat([fuzzy_data, dos_data], ignore_index=True)
```

```python
combined_data.to_csv('combined_data.csv', index=False)
```

```python
combined_data['Class_Instance'].value_counts()
```

```python
len(combined_data)
```

```python
import seaborn as sns
plt.figure(figsize =(10,4), dpi = 50)
sns.countplot(data=combined_data, x ='Class_Instance');
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize label encoder
label_encoder = LabelEncoder()

# Encode label column
combined_data['Class_Instance'] = label_encoder.fit_transform(combined_data['Class_Instance'])

# Verify the transformation 
print(combined_data['Class_Instance'].unique())  # Check unique values after encoding
```

```python
combined_data
```

```python
combined_data['Class_Instance'].value_counts()
```

```python
combined_data.to_csv('combined_data_encodedFinal.csv', index = False)
```

```python
#Import machine learning models 
from sklearn.metrics import confusion_matrix, matthews_corrcoef
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
```

```python
#Split data into train_test Split 
# Split data into features (X) and labels (y)

X = combined_data.drop(columns=['Class_Instance','DLC', 'Timestamp'])

y = combined_data['Class_Instance']
```

```python
combined_data.to_csv('combined_data_encoded.csv')
```

```python
#split traiing and test 70% to 30% 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
len(X_train)
```

```python
y_test
```

```python

# Initialize and train models
rf_model = RandomForestClassifier(n_estimators= 10,max_features ='sqrt',
                                                random_state =101)
rf_model.fit(X_train, y_train)

dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)

xgboost_model = XGBClassifier(objective='binary:logistic', random_state=42)
xgboost_model.fit(X_train, y_train)

knn_model = KNeighborsClassifier(n_neighbors=1)
knn_model.fit(X_train, y_train)
```

```python
import joblib

# Saved trained models
joblib.dump(rf_model, 'random_forest_model.pkl')
joblib.dump(dt_model, 'decision_tree_model.pkl')
joblib.dump(xgboost_model, 'xgboost_model.pkl')
joblib.dump(knn_model, 'knn_model.pkl')
```

```python
# Make predictions and evaluate models
rf_predictions = rf_model.predict(X_test)
dt_predictions = dt_model.predict(X_test)
xgboost_predictions = xgboost_model.predict(X_test)
knn_predictions = knn_model.predict(X_test)
```

```python
# Calculate metrics
rf_mcc = matthews_corrcoef(y_test, rf_predictions)
dt_mcc = matthews_corrcoef(y_test, dt_predictions)
xgboost_mcc = matthews_corrcoef(y_test, xgboost_predictions)
knn_mcc = matthews_corrcoef(y_test, knn_predictions)

rf_cm = confusion_matrix(y_test, rf_predictions)
dt_cm = confusion_matrix(y_test, dt_predictions)
xgboost_cm = confusion_matrix(y_test, xgboost_predictions)
knn_cm = confusion_matrix(y_test, knn_predictions)

# Print or use these metrics for your comparative analysis
print("Random Forest MCC:", rf_mcc)
print("Decision Tree MCC:", dt_mcc)
print("XGBoost MCC:", xgboost_mcc)
print("KNN MCC:", knn_mcc)
print("Confusion Matrix (Random Forest):\n", rf_cm)
print("Confusion Matrix (Decision Tree):\n", dt_cm)
print("Confusion Matrix (XGBoost):\n", xgboost_cm)
print("Confusion Matrix (KNN):\n", knn_cm)
```

```python
algorithms = ['Random Forest', 'Decision Tree', 'XGBoost', 'KNN']
confusion_matrices = []

for model in [rf_model, dt_model, xgboost_model, knn_model]:
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    confusion_matrices.append(cm)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
for i, ax in enumerate(axes.flatten()):
    sns.heatmap(confusion_matrices[i], annot=True, fmt='d', cmap='Blues', ax=ax)
    ax.set_title(algorithms[i])
    ax.set_xlabel('Predicted Label')
    ax.set_ylabel('True Label')
plt.tight_layout()
plt.show()
```

```python
#One model at a time performance evaluation
#Random Forest
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(y_test, rf_predictions)
print(classification_report(y_test, rf_predictions))
classes = ['Denial of Service', 'Fuzzing Attack', 'Normal Message']
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes)
disp.plot(cmap='viridis', values_format='d')
```

```python
model_name = 'Random Forest'

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['DoS', 'Fuzzing', 'Attack Free'])
disp.plot(cmap=plt.cm.Blues, values_format='d')
plt.title(f'Confusion Matrix for {model_name}')
plt.show()
```

```python
class_labels = ['DoS', 'Fuzzing', 'Attack Free']

# TP, TN, FP, FN for DoS, Fuzzing and Normal message 
for i in range(len(class_labels)):
    class_name = class_labels[i]
    tp = cm[i, i]
    fn = np.sum(cm[i, :]) - tp
    fp = np.sum(cm[:, i]) - tp
    tn = np.sum(cm) - (tp + fn + fp)
    
    # Print metrics for the current class
    print(f"Metrics for {class_name}:")
    print(f"True Positive: {tp}")
    print(f"True Negative: {tn}")
    print(f"False Positive: {fp}")
    print(f"False Negative: {fn}")
    print("-" * 30)
```

```python
#Calculate Detection Times for RANDOM FOREST
#Estimate True psotive indexes for True postives for both DoS and Fuzzing in the attack set 
#These are indicies where DoS and Fuzzing was correctly predicted
true_positive_indices_dos = [i for i in range(len(rf_predictions)) if rf_predictions[i] == 0]  # 0 Represents DoS class in my test Data X_test
true_positive_indices_fuzzing = [i for i in range(len(rf_predictions)) if rf_predictions[i] == 1]  # 1 Represents Fuzzing class in X_test
true_positive_indices = true_positive_indices_dos + true_positive_indices_fuzzing
detection_times = []
start_time = time.time()
rf_predictions = rf_model.predict(X_test) # Predict for the specific instance
end_time = time.time()
detection_times.append(end_time - start_time) #calculate and store detection time
#test Time
prediction_time = end_time - start_time
Average_detection_time = sum(detection_times) / len(true_positive_indices)
print(f"prediction_time: {prediction_time} seconds")
print(f"Average Detection Time for each TPs: {Average_detection_time} seconds")
```
